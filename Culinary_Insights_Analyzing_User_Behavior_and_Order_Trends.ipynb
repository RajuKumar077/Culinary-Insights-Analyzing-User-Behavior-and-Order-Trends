# Merging Logic
The three datasets provided (UserDetails, CookingSessions, and OrderDetails) have certain common columns that can be used for merging:

**UserDetails and CookingSessions:** Both datasets share the User ID column, which uniquely identifies each user. By merging these two on User ID, we can associate each user’s personal details with their cooking sessions, i.e., we can see which dishes a user has cooked, what session they were involved in, and more.

**UserDetails and OrderDetails:** Similarly, the User ID column is also present in OrderDetails, allowing us to link each user’s details with their order history. This will allow us to understand which dishes a user has ordered, how many orders they’ve made, and what meal type they prefer.

**CookingSessions and OrderDetails:** The key column here is Session ID, which is shared between CookingSessions and OrderDetails. This enables us to link each cooking session to the corresponding order, which is important for understanding the relationship between what was cooked and what was ordered, and when these events took place.

```python
user_cooking_data = pd.merge(user_details, cooking_sessions, on='User ID', how='inner')
```

```python
full_data = pd.merge(user_cooking_data, order_details, on='Session ID', how='inner')
```

I merged the three datasets—`UserDetails`, `CookingSessions`, and `OrderDetails`—using common columns: `User ID` for linking user details with cooking sessions and orders, and `Session ID` for connecting cooking sessions with orders. The inner join ensures that only records with matching data are included. This creates a unified dataset to analyze user demographics, cooking habits, and order trends.

| **Step**                       | **Datasets Merged**                    | **Merge Key**     | **Purpose**                                              |
|---------------------------------|----------------------------------------|-------------------|----------------------------------------------------------|
| **Step 1**                     | `UserDetails` + `CookingSessions`      | `User ID`         | Link user info with their cooking sessions               |
| **Step 2**                     | Result from Step 1 + `OrderDetails`    | `Session ID`      | Connect cooking sessions with corresponding order details |
| **Final Dataset**               | `UserDetails`, `CookingSessions`, `OrderDetails` | `User ID`, `Session ID` | Unified dataset for analyzing user behavior, cooking, and orders |

```python
import pandas as pd
import plotly.graph_objects as go



df = pd.DataFrame(full_data)

# 1. Age Distribution
age_fig = go.Figure()
age_fig.add_trace(go.Histogram(x=df['Age'], nbinsx=10, name="Age Distribution"))
age_fig.update_layout(title="Age Distribution of Users", xaxis_title="Age", yaxis_title="Count")
```

```python

# 2. Favorite Meal Distribution
meal_fig = go.Figure()
meal_fig.add_trace(go.Pie(labels=df['Favorite Meal'], values=df['Total Orders'], hole=0.3, name="Favorite Meals"))
meal_fig.update_layout(title="Favorite Meals by Total Orders")


```

```python
# 3. Order Trend Over Time
df['Order Date'] = pd.to_datetime(df['Order Date'])
order_trend = df.groupby('Order Date')['Amount (USD)'].sum().reset_index()
order_trend_fig = go.Figure()
order_trend_fig.add_trace(go.Scatter(x=order_trend['Order Date'], y=order_trend['Amount (USD)'], mode='lines+markers', name="Order Trend"))
order_trend_fig.update_layout(title="Order Trends Over Time", xaxis_title="Order Date", yaxis_title="Total Amount (USD)")

```

```python
import plotly.graph_objects as go

# Categorize orders into Completed and Not Completed
order_status_summary = df['Order Status'].value_counts()
completed = order_status_summary.get('Completed', 0)
not_completed = order_status_summary.sum() - completed

# Create a pie chart
order_completion_fig = go.Figure(
    go.Pie(
        labels=['Completed', 'Not Completed'],
        values=[completed, not_completed],
        hole=0.4  # Makes it a donut chart for better aesthetics
    )
)

order_completion_fig.update_layout(
    title="Order Completion Status",
    annotations=[
        dict(
            text="Order Status",
            x=0.5,
            y=0.5,
            font_size=15,
            showarrow=False
        )
    ]
)

order_completion_fig.show()
```

```python

# 5. Amount vs Total Orders (Scatter Plot)
amount_vs_orders_fig = go.Figure()
amount_vs_orders_fig.add_trace(go.Scatter(x=df['Total Orders'], y=df['Amount (USD)'], mode='markers', text=df['User Name'], name="Amount vs Orders"))
amount_vs_orders_fig.update_layout(title="Order Amount vs Total Orders", xaxis_title="Total Orders", yaxis_title="Amount (USD)")

```

```python
# 6. Session Duration by User
session_duration_fig = go.Figure()
session_duration_fig.add_trace(go.Bar(x=df['User Name'], y=df['Duration (mins)'], name="Session Duration", marker=dict(color='blue')))
session_duration_fig.update_layout(title="Session Duration by User", xaxis_title="User Name", yaxis_title="Duration (mins)")

```

```python
# 7. Session Rating by Meal Type (Box Plot)
session_rating_fig = go.Figure()
session_rating_fig.add_trace(go.Box(x=df['Meal Type'], y=df['Session Rating'], name="Session Rating"))
session_rating_fig.update_layout(title="Session Ratings by Meal Type", xaxis_title="Meal Type", yaxis_title="Rating")

```
